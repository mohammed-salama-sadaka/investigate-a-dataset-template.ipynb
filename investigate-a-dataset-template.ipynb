## Exploratory Data Analysis

### Research Question 1 (Does the body weight affects sprint speed and stamina for a player?)

Merge players dataframe with players attributes on player_api_id column.

```python
dataframe = pd.merge(players[['player_api_id','weight']] , players_attributes[['player_api_id','stamina','balance','sprint_speed','strength','short_passing','standing_tackle','dribbling','ball_control','shot_power','long_passing','vision','defensive_work_rate','attacking_work_rate']], on= ['player_api_id'])
dataframe.head()
```

Aggregate data based on the average recorded attributes values over years.

```python
dataframe__= dataframe.groupby('player_api_id', as_index= False).mean()
```

```python
dataframe__.plot(x= "weight", y= "balance", kind= "scatter", figsize= (8,8), title= "Correlation between body weight and balance.")
```

```python
dataframe__.plot(x= "weight", y= "sprint_speed",kind= "scatter", figsize= (8,8), title= "Correlation between body weight and sprint speed.")
```

```python
dataframe__.plot(x= "weight", y= "stamina", kind= "scatter", figsize= (8,8), title= "Correlation between body weight and stamina.")
```

### Research Question 2  (How to determine if a player is a defender or attacker based on player attributes?)

Determine if a player is a defender or an attacker. I will drop players that arn't neither defenders nor attackers.

```python
def classify_player( row ):
    if row['defensive_work_rate'] == 'high':
        return 'defender'
    elif row['attacking_work_rate'] == 'high':
        return 'attacker'
    else:
        return 'no'
dataframe['category'] = dataframe.apply( classify_player, axis= 1 )
```

```python
dataframe = dataframe.drop(dataframe[dataframe['category'] == 'no'].index)
```

Drop player_api_id column from my dataframe because it's not needed for answering the second question.

```python
dataframe.drop(columns= dataframe[['player_api_id','defensive_work_rate','attacking_work_rate','weight']],inplace= True)
```

Calculate the average value for each player attribute for all defender and attacker players in the dataset.

```python
dataframe_= dataframe.groupby(['category'], as_index= True).mean()
```

```python
dataframe_.transpose().plot(kind= "bar", figsize= (12,8), title= "avg player attributes for defender", color= ['darkgreen','lightgreen'])
```
